```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('/kaggle/input/250k-medicines-usage-side-effects-and-substitutes/medicine_dataset.csv',
                 low_memory=False)

# Filter only habit forming medicines
habit_forming_df = df[df["Habit Forming"].str.lower() == "yes"]

# Count by chemical class
habit_counts = habit_forming_df["Chemical Class"].value_counts().head(10)

print("Top 10 Chemical Classes forming habits:")
print(habit_counts)

# Visualization
plt.figure(figsize=(10,6))
habit_counts.plot(kind='bar')
plt.title("Top 10 Chemical Classes Forming Habits")
plt.ylabel("Number of Medicines")
plt.xticks(rotation=45, ha="right")
plt.savefig("/kaggle/working/top_habit_forming_classes.png", bbox_inches='tight', dpi=300)
plt.show()
```
